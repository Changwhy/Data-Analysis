## Woman_Child_Groups

```python
train.head(1)
```

```python
def extract_surname(name):
    return name.split(',')[0]

def ticket_X(ticket):
    return ticket[:-1] + 'X'
```

```python
train['Surname'] = train['Name'].apply(extract_surname)

train['Ticket_X'] = train['Ticket'].apply(ticket_X)

train['GroupId'] = train['Surname']+ train['Pclass'].astype(str) + train['Embarked'] + train['Ticket_X']
```

```python
train.head()
```

```python
train['Name_title'] = train['Name'].str.extract('([A-Za-z]+)\.')
```

```python
test['Name_title'] = test['Name'].str.extract('([A-Za-z]+)\.')
```

```python
newtitles={
    "Capt":       "Mr",
    "Col":        "Mr",
    "Major":      "Mr",
    "Jonkheer":   "Mr",
    "Don":        "Mr",
    "Sir" :       "Mr",
    "Dr":         "Mr",
    "Rev":        "Mr",
    "Dona":       "Mrs",
    "Mme":        "Mrs",
    "Mlle":       "Miss",
    "Ms":         "Mrs",
    "Mr" :        "Mr",
    "Mrs" :       "Mrs",
    "Miss" :      "Miss",
    "Master" :    "Master",
    "Lady" :      "Mrs",
    "Countess" : "Mrs"}

train['Name_title'] = train['Name_title'].map(newtitles)
test['Name_title'] = test['Name_title'].map(newtitles)
```

```python
train.head()
```

```python
train.Name_title.value_counts()
```

```python
train['GroupId_Freq'] = train.groupby('GroupId')['GroupId'].transform('count')
```

```python
train.loc[train['Name_title']=='Mr', 'GroupId'] = 'X'
train.loc[train['GroupId_Freq']<=1, 'GroupId'] = 'X'
```

```python
GroupSurvival= train.groupby('GroupId')['Survived'].mean()
```

```python
train['GroupId_Survival'] = train['GroupId'].map(dict(GroupSurvival))
```

```python
GroupSurvival.value_counts()
```

```python
train.loc[train['GroupId_Survival']==1, 'pred'] = 1
train.loc[train['GroupId_Survival']==0, 'pred'] = 0
```

```python
train.loc[(train['GroupId']=='X')&(train['Sex']=='male'), 'pred'] = 0
train.loc[(train['GroupId']=='X')&(train['Sex']=='female'), 'pred'] = 1
```
